```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
import statsmodels.formula.api as smf

df = pd.read_csv('Worlds Best Employers.csv')
df.head(5)
```

```python
df.describe()
```

```python
df2 = pd.DataFrame(df['COUNTRY/TERRITORY'].value_counts()).reset_index()
df2 = df2.rename(columns = {'index': 'Country', 'COUNTRY/TERRITORY': 'Count'})
df2
```

```python
bars = alt.Chart(df2).mark_bar().encode(
    x = alt.X('Count:Q'),
    y = alt.Y('Country:N', sort = alt.EncodingSortField(field = "Country", op = "count", order = 'ascending')),
    color = alt.Color('Country', sort = '-x', legend = None)
)

text = bars.mark_text(
    align = 'left',
    baseline = 'middle',
    dx = 3  # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text = 'Count:Q'
)

(bars + text).properties(height = 400, width = 800)
```

```python
df3 = pd.DataFrame(df['INDUSTRIES'].value_counts()).reset_index()
df3 = df3.rename(columns = {'index': 'Industry', 'INDUSTRIES': 'Count'})
df3
```

```python
bars = alt.Chart(df3).mark_bar().encode(
    x = alt.X('Count:Q'),
    y = alt.Y('Industry:N', sort = alt.EncodingSortField(field = 'Industry', op = 'count', order = 'ascending')),
    color = alt.Color('Industry', sort = '-x', legend = None)
)

text = bars.mark_text(
    align = 'left',
    baseline = 'middle',
    dx = 3  # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text = 'Count:Q'
)

(bars + text).properties(height = 400, width = 800)
```

```python
base = alt.Chart(df3).mark_arc(innerRadius = 50).encode(
    theta = alt.Theta(field = 'Count', type = 'quantitative'),
    color = alt.Color(field = 'Industry', type = 'nominal', sort = alt.EncodingSortField(field = 'Industry', op = 'count', order = 'descending')),
    order = alt.Order(field = 'Count', type = "quantitative", sort = 'ascending')
)

text = alt.Chart(df3).mark_text(radius = 120).encode(
    theta = alt.Theta('Count:Q', stack = True),
    text = 'Count'
)

(base + text).properties(height = 400, width = 800)
```
